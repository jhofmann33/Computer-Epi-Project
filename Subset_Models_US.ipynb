```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
```

```python
# Import the Excel file into a pandas dataframe
df_2015 = pd.read_excel('US_2015_influenza_Data.xlsx', engine='openpyxl')
df_2015 = df_2015.sort_values(by='ISO_SDATE', ascending=True)
#Display data
# display(df)
```

```python
# Group the data by ISO_WEEK 
df_agg = df_2015.groupby('ISO_WEEK').agg({'INF_ALL': 'mean', 'INF_A': 'mean', 'INF_B': 'mean'}).reset_index()
#InF_ALL = INF_A + INF_B cases.
#display aggregate
# display(df_agg)
```

# US 2015 Influenza Cases

```python
# Plot INF_ALL, INF_A, and INF_B on the same graph
plt.plot(df_agg['ISO_WEEK'], df_agg['INF_ALL'], label='Cumulative Cases')
plt.plot(df_agg['ISO_WEEK'], df_agg['INF_A'], label='Influenza A Cases')
plt.plot(df_agg['ISO_WEEK'], df_agg['INF_B'], label='Influenza B Cases')
plt.xlabel('Week')
plt.ylabel('Number of Cases')
plt.title('US - 2015 - Influenza Cases by Week')
plt.legend()
plt.show()
```
