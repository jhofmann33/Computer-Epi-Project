```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import the Excel file into a pandas dataframe
df_2020 = pd.read_excel('China_2020_Covid19_Data.xlsx', engine='openpyxl')
df_2020 = df_2020.sort_values(by='ISO_START_DATE', ascending = True)
# display(df_2020)
```

```python
df_agg = df_2020.groupby(['ISO_YEAR','ISO_START_DATE']).agg({'DAILY_CASES': 'mean'}).reset_index()
#Cleaning the string to not show full date data like time etc.
df_agg['ISO_START_DATE'] = df_agg['ISO_START_DATE'].str.split('T').str[0]
# display(df_agg)
```

# China 2020 Covid Cases

```python
#Create a line plot 
plt.figure(figsize=(16,8))
#Don't know why but the data gets normalized have to divide by 10 to show true numbers
plt.bar(df_agg['ISO_START_DATE'], df_agg['DAILY_CASES'], label='Covid Cases')

plt.xlabel('DATE')
plt.ylabel('Number of Cases')
plt.title('China - COVID19 cases 2020')

xtick_pos = np.arange(0, len(df_agg['ISO_START_DATE']), 5)
plt.xticks(xtick_pos, df_agg['ISO_START_DATE'][xtick_pos], rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
# Load the data from the Excel file
df_2020_to_2022 = pd.read_excel('China_2020-2022_Covid19_Data.xlsx', engine='openpyxl')
df_agg4 = df_2020_to_2022.groupby(['ISO_YEAR','ISO_START_DATE']).agg({'DAILY_CASES': 'mean'}).reset_index()
df_agg4['ISO_START_DATE'] = df_agg4['ISO_START_DATE'].str.split('T').str[0]
plt.figure(figsize=(20, 10))
plt.bar(df_agg4['ISO_START_DATE'], df_agg4['DAILY_CASES'], label='Covid Cases')
plt.xlabel('DATE')
plt.ylabel('Number of Cases')
plt.title('China - COVID19 cases 2020 - 2022')

xtick_pos = np.arange(0, len(df_agg4['ISO_START_DATE']), 10)
plt.xticks(xtick_pos, df_agg4['ISO_START_DATE'][xtick_pos], rotation=45)
plt.legend()
plt.grid()
plt.show()
```
