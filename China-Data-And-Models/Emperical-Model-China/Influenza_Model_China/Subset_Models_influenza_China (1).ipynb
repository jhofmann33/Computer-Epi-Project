```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Import the Excel file into a pandas dataframe
df_2015 = pd.read_excel('China_2015_influenza_Data.xlsx', engine='openpyxl')
df_2015 = df_2015.sort_values(by='ISO_SDATE', ascending=True)
```

```python
# Group the data by year and ISO_WEEK, calculate the sum (sentinel and non-sentinel) of INF_ALL for each group
df_agg = df_2015.groupby(['ISO_YEAR', 'ISO_SDATE']).agg({'INF_ALL': 'sum', 'INF_A': 'sum', 'INF_B': 'sum'}).reset_index()
# Extract only the date portion of ISO_SDATE column
df_agg['ISO_SDATE'] = df_agg['ISO_SDATE'].str.split('T').str[0]
```

# China 2015 Influenza cases 

```python
# Create a line plot using ISO_WEEK as the x-axis and INF_ALL as the y-axis
plt.figure(figsize=(16, 6))
plt.plot(df_agg['ISO_SDATE'], df_agg['INF_ALL'], label='Cumulative Cases')
plt.plot(df_agg['ISO_SDATE'], df_agg['INF_A'], label='Influenza A Cases')
plt.plot(df_agg['ISO_SDATE'], df_agg['INF_B'], label='Influenza B Cases')

plt.xlabel('DATE')
plt.ylabel('Number of Cases')
plt.title('China - Influenza Cases from 2015')

xtick_pos = np.arange(0, len(df_agg['ISO_SDATE']), 5)
plt.xticks(xtick_pos, df_agg['ISO_SDATE'][xtick_pos], rotation=45, fontsize=12)
plt.grid()
plt.legend()
plt.show()
```

# China 2020-2022 influenza cases with specimine collection

```python
# Load the data from the Excel file
df_2020_to_2022 = pd.read_excel('China_2020-2022_influenza_Data.xlsx', engine='openpyxl')

# Group the data by year and ISO_WEEK, calculate the sum (sentinel and non-sentinel) of INF_ALL for each group
df_agg11 = df_2020_to_2022.groupby(['ISO_YEAR', 'ISO_SDATE']).agg({'INF_ALL': 'sum', 'INF_A': 'sum', 'INF_B': 'sum'}).reset_index()

df_agg_samples = df_2020_to_2022.groupby(['ISO_YEAR','ISO_SDATE']).agg({'SPEC_PROCESSED_NB' :'sum'}).reset_index()
# # Extract only the date portion of ISO_SDATE column
df_agg11['ISO_SDATE'] = df_agg11['ISO_SDATE'].str.split('T').str[0]
df_agg_samples['ISO_SDATE'] = df_agg_samples['ISO_SDATE'].str.split('T').str[0]
# Create a figure with two subplots
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(16, 16))

# Plot the line graphs in the first subplot
ax1.plot(df_agg11['ISO_SDATE'], df_agg11['INF_ALL'], label='Cumulative Cases')
ax1.plot(df_agg11['ISO_SDATE'], df_agg11['INF_A'], label='Influenza A Cases')
ax1.plot(df_agg11['ISO_SDATE'], df_agg11['INF_B'], label='Influenza B Cases')
ax1.set_xlabel('DATE')
ax1.set_ylabel('Number of Cases')
ax1.set_title('China - Influenza Cases from 2020 to 2022')
xtick_pos = np.arange(0, len(df_agg11['ISO_SDATE']), 10)
ax1.set_xticks(xtick_pos)
ax1.set_xticklabels(df_agg11['ISO_SDATE'][xtick_pos], rotation=45)
ax1.legend()
ax1.grid()

# Plot the bar graph in the second subplot
ax2.bar(df_agg_samples['ISO_SDATE'], df_agg_samples['SPEC_PROCESSED_NB'], label='Samples Process')
ax2.set_xlabel('DATE')
ax2.set_ylabel('Number of Samples')
ax2.set_title('China - Samples Processed from 2020 to 2022')
xtick_pos = np.arange(0, len(df_agg_samples['ISO_SDATE']), 10)
ax2.set_xticks(xtick_pos)
ax2.set_xticklabels(df_agg_samples['ISO_SDATE'][xtick_pos], rotation=45)
ax2.legend()
ax2.grid()

# Adjust the spacing between the subplots
fig.subplots_adjust(hspace=0.4)

# Show the combined graph with both line and bar plots
plt.show()
```
